```python
import matplotlib.pyplot as plt
def visualize_genre_distribution(movies_df):
    genre_counts = movies_df['genres'].str.split('|').explode().value_counts()
    plt.figure(figsize=(10, 6))
    genre_counts.plot(kind='bar')
    plt.title('Movie Genres Distribution')
    plt.xlabel('Genres')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
import seaborn as sns
def showmp(ratings_df):
    plt.figure(figsize=(8, 6))
    sns.histplot(ratings_df['rating'], bins=10, kde=True)
    plt.title('User Ratings Distribution')
    plt.xlabel('Rating')
    plt.ylabel('Count')
    plt.show()
    
showmp(ratings_df)
```

```python
def showmp1(ratings_df):
    ratings_df['timestamp'] = pd.to_datetime(ratings_df['timestamp'], unit='s')
    ratings_df['year'] = ratings_df['timestamp'].dt.year
    average_ratings_over_time = ratings_df.groupby('year')['rating'].mean()
    plt.figure(figsize=(10, 6))
    average_ratings_over_time.plot()
    plt.title('Average Movie Ratings Over Time')
    plt.xlabel('Year')
    plt.ylabel('Average Rating')
    plt.grid(True)
    plt.show()
showmp1(ratings_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def showmp2(user_item_matrix):
    plt.figure(figsize=(10, 6))
    sns.heatmap(user_item_matrix, cmap='viridis')
    plt.title('User-Item Interaction Heatmap')
    plt.xlabel('Item ID')
    plt.ylabel('User ID')
    plt.show()
showmp2(user_item_matrix)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
item_user_matrix = user_item_matrix.T
item_similarity = cosine_similarity(item_user_matrix, dense_output=False)
np.fill_diagonal(item_similarity, 0)
item_similarity_df = pd.DataFrame(item_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)
print(item_similarity_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def showmp3(item_similarity_df):
    plt.figure(figsize=(10, 8))
    sns.heatmap(item_similarity_df, cmap='coolwarm', vmin=0, vmax=1)
    plt.title('Item-Item Similarity Matrix')
    plt.xlabel('Item ID')
    plt.ylabel('Item ID')
    plt.show()

showmp3(item_similarity_df)

```

```python
#Fuction for collaborative filtering recommendations
```

```python


def get_top_similar_items(item_similarity_df, item_id, top_n=5):
    similar_items_row = item_similarity_df.loc[item_id]
    top_similar_items = similar_items_row.sort_values(ascending=False).head(top_n)
    return top_similar_items



```

```python
def generate_recommendations(user_id, user_item_matrix, item_similarity_df, top_n=5):
    user_items = user_item_matrix.loc[user_id]

    item_scores = {}
    for item_id, interaction in user_items.items():
        similar_items = get_top_similar_items(item_similarity_df, item_id, top_n)
        
        for similar_item_id, similarity_score in similar_items.items():
            if similar_item_id not in item_scores:
                item_scores[similar_item_id] = 0
            item_scores[similar_item_id] += similarity_score * interaction
    
    sorted_item_scores = sorted(item_scores.items(), key=lambda x: x[1], reverse=True)
    recommendations = [item_id for item_id, score in sorted_item_scores[:top_n]]
    
    return recommendations
```

```python
user_id = 1
recommendations = generate_recommendations(user_id, user_item_matrix, item_similarity_df)
print("Recommendations for entered user", user_id, ":", recommendations)
```

```python
# Content Based Recommendations on "genre"
```

```python
def recommend_movies_by_genre(user_id, final_table, top_n=5):
    user_data = final_table[final_table['userId'] == user_id]

    genre_counts = {}
    for genres in user_data['genres']:
        for genre in genres.split('|'):
            if genre in genre_counts:
                genre_counts[genre] += 1
            else:
                genre_counts[genre] = 1
    sorted_genres = sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)
    top_genres = [genre for genre, _ in sorted_genres[:top_n]]

    recommended_movies = final_table[~final_table['title'].isin(user_data['title']) & 
                                     final_table['genres'].apply(lambda x: any(genre in x.split('|') for genre in top_genres))]
    recommended_movies = recommended_movies[~recommended_movies['title'].isin(user_data['title'])]

   
    recommended_movies = recommended_movies.drop_duplicates(subset=['title'])

    return recommended_movies
```

```python


user_id = 1
recommended_movies = recommend_movies_by_genre(user_id, final_table)
print("Recommendations for user", user_id, ":")
print(recommended_movies[['title', 'genres']].head())
```
